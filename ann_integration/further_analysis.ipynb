# Marker analysis for unknown population annotation

```python
import pandas as pd
import numpy as np
import scanpy as sc
import anndata as ad
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Define locations
path_to_adata = "../data/processed/combined_annotated.h5ad"
output_dir = "../results/ann_integration/further_analysis/"
```

Assuming that your adata has the final label prediction in adata.obs['final_pred']

```python
adata = sc.read(path_to_adata)
```

```python
marker_genes = {}
for label in adata.obs['final_pred'].unique():
    print(f"Analyzing marker genes for label: {label}")
    marker_genes[label] = {}
    sc.tl.rank_genes_groups(adata, 'final_pred', groups=[label], reference='rest', method='wilcoxon')
    sc.pl.rank_genes_groups(adata, n_genes=25, sharey=False, save=f"_{label}_marker_genes.png")
```
